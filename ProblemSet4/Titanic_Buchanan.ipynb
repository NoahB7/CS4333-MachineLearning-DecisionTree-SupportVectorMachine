```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn import svm
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("Titanic.txt")
X = data.drop("SURVIVED", axis=1)
y = data.pop("SURVIVED")
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
y_test = np.array(y_test).astype(int)
```

# Random Forest experimentation

```python
for i in range(1,7):
    for j in range(1,7):
        rf = RandomForestClassifier(max_features=i, max_depth=j, random_state=0)
        rf.fit(X_train,y_train)
        yhat = rf.predict(X_test)
        correct = 0;
        for h in range(len(yhat)):
            if y_test[h] == yhat[h]:
                correct += 1

        print(correct/len(yhat))
    print(",")

```

basline: 0.7506

# Decision Tree experimentation

```python
for i in range(1,7):
    for j in range(1,7):
        dt = DecisionTreeClassifier(max_features = i, max_depth = j,random_state=0)
        dt.fit(X_train,y_train)
        yhat = dt.predict(X_test)
        correct = 0;
        for h in range(len(yhat)):
            if y_test[h] == yhat[h]:
                correct += 1

        print(correct/len(yhat))
    print(",")
```

baseline: 0.7343

# Support Vector Machine experimentation

```python
kernels = ['linear', 'poly', 'rbf', 'sigmoid']
for i in range(len(kernels)):
    for j in range(1,11):
        svc = svm.SVC(kernel = kernels[i],C = j*.2)
        svc.fit(X_train, y_train)
        yhat = svc.predict(X_test)
        correct = 0;
        for h in range(len(yhat)):
            if y_test[h] == yhat[h]:
                correct += 1

        print(correct/len(yhat))
    print(",")
```

baseline: 0.7902

# Nearest Neighbors experimentation

```python
weights = ['uniform', 'distance']
for i in range(1,11):
    for j in range(2):
        neigh = KNeighborsClassifier(n_neighbors=i, weights=weights[j])
        neigh.fit(X_train, y_train)
        yhat = neigh.predict(X_test)
        correct = 0;
        for h in range(len(yhat)):
            if y_test[h] == yhat[h]:
                correct += 1

        print(correct/len(yhat))
    print(",")
```

baseline: 0.7599
